```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import numpy as np
import pandas as pd
```

```python
## load the ann scalar and ohe pickles
model=load_model('model.h5')

with open('Ohe_geographpy.pkl', 'rb') as file:
    ohe=pickle.load(file)

with open('Lable_encoder_gender.pkl', 'rb') as file:
    lable_encoder=pickle.load(file)
  
with open('scalar.pkl','rb') as file:
    scalar=pickle.load(file)
    
```

```python
input_data = {
    "CreditScore" : 600,
    "Geography": "France",
    "Gender" : "Male",
    "Age" : 40,
    "Tenure" : 3,
    "Balance" : 60000,
    "NumOfProducts" : 2,
    "HasCrCard": 1,
    "IsActiveMember": 1,
    "EstimatedSalary" : 50000
}
```

```python
geo_encoded=ohe.transform([[input_data["Geography"]]])
geo_encoded_df=pd.DataFrame(geo_encoded,columns=ohe.get_feature_names_out())
geo_encoded_df
```

```python
input_df=pd.DataFrame([input_data])
input_df.drop(columns=['Geography'],axis=1,inplace=True)
input_df
```

```python
df=pd.concat([input_df,geo_encoded_df],axis=1)
```

```python
df
```

```python
df['Gender']=lable_encoder.transform(df['Gender'])
df
```

```python
df=scalar.transform(df)
```

```python
df
```

```python
prediction=model.predict(df)
```
